<h1>Problem 1: Part 1</h1>

```python
def preceding(char):
    if char == 'a':
        return 'z'
    elif char == 'A':
        return 'Z'
    elif (char >= 'B' and char <= 'Z') or (char >= 'b' and char <= 'z'):
        return chr(ord(char)-1)
    else:
        return char
```

```python
# Check:
print(preceding('a')) #Should print z
print(preceding('A')) #Should print Z

print(preceding('b')) #Should print a
print(preceding('c')) #Should print b
print(preceding('z')) #Should print y

print(preceding('B')) #Should print A
print(preceding('C')) #Should print B
print(preceding('Z')) #Should print Y

print(preceding('!')) #Should print !
print(preceding('')) #Should print ''
```

<h1>Problem 1: Part 2</h1>


```python
def succeeding(char):
    if char == 'z':
        return 'a'
    elif char == 'Z':
        return 'A'
    elif (char >= 'A' and char <= 'Y') or (char >= 'a' and char <= 'y'):
        return chr(ord(char)+1)
    else:
        return char   
```

```python
# Check: 
print(succeeding('z')) #Should print a
print(succeeding('Z')) #Should print A

print(succeeding('A')) #Should print B
print(succeeding('B')) #Should print C
print(succeeding('Y')) #Should print Z

print(succeeding('a')) #Should print b
print(succeeding('b')) #Should print c
print(succeeding('y')) #Should print z

print(succeeding('!')) #Should print !
print(succeeding(' ')) #Should print ' '
```

<h1>Problem 1: Part 3</h1>

```python
def message_coder(message_string,function):
    coded_message = ''
    for c in message_string:
        coded_message += function(c)
    return coded_message
```

```python
# Check:
print(message_coder('hello dolly zebra!',preceding)) # Should print 'gdkkn cnkkx ydaqz!’

print(message_coder('hello dolly zebra!',succeeding)) # Should print 'ifmmp epmmz afcsb!’
```

<h1>Problem 2: Word Count</h2>

```python
def word_distribution(text_string,word_list=[]):
    frequency_dict = dict()
    
    if word_list:
        for word in word_list:
            frequency_dict[word.lower()] = 0

    for word in text_string.split():
        word = word.lower()
        
        while len(word)>=1 and not (word[-1] >= 'a' and word[-1] <= 'z'):
            word = word[:-1]
            
        if word == '':
            continue
             
        if(word in frequency_dict or not word_list):
            if frequency_dict.get(word) is None:
                frequency_dict[word] = 1
            else:
                frequency_dict[word] += 1
    
    return frequency_dict
```

```python
# Check: 
# (1)
text_string = "Hello. How are you? Please say hello if you don’t love me!"
print(word_distribution(text_string)) 
# should print {‘hello’: 2, ‘how’:1, ‘are’:1, ‘you’:2,’please’:1, “don’t”: 1 …}

# (2)
text_string = 'I came, I saw, I conquered!'
print(word_distribution(text_string,word_list=['I','saw','Britain'])) 
# should print {‘i’:3,’saw’:1,’britain':0}
```
